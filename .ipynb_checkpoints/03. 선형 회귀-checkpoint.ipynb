### 선형 회귀
- 독립 변수를 사용해 종속 변수의 움직임을 예측하고 설명하는 작업
- 쉽게 풀어 얘기하면 가장 훌륭한 예측선 긋기
- 하나의 독립 변수만으로도 종속 변수값을 설명할 수 있을 때 이를 <b>단순 선형 회귀</b>(simple linear regression)라고 한다
- 독립 변수가 여러 개 필요할 때 <b>다중 선형 회귀</b>(multiple linear regression)라고 한다

### cost(Mean Square error) 

#### cost 계산

```python
# 독립 변수
x_data = np.array([1, 2, 3])
# 종속 변수
y_data = np.array([1, 2, 3])
```

```python
plt.plot(x_data, y_data)
plt.show()
```

x_data와 y_data에 대해 라인 그래프를 그리면 기울기가 1인 선이 그려진다.

```python
# cost 구하는 함수
def cost(x, y, w): # 독립 변수, 종속 변수, 기울기(가중치)
    # 예측값
    hx = w * x
    # cost값
    c = np.mean((hx - y) ** 2)

    return c
```

```python
# 기울기를 -1이라고 설정햇을 때 cost값
cost(x_data, y_data, -1)
```

엉뚱하게 예측해 예측값과 실제값이 차이가 많이 나기 때문에 cost값이 크게 나온다

```python
# 기울기를 1이라고 설정했을 때 cost값
cost(x_data, y_data, 1)
```

예측값과 실제값이 같기 때문에 cost값은 0

#### 기울기(가중치)에 따른 cost값 

```python
for w in np.linspace(-3, 5, 50): # -3부터 5까지 50등분 
    # cost값
    c = cost(x_data, y_data, w)
    # 기울기에 대한 cost값 scatter 그래프
    plt.plot(w, c, 'ro')
plt.xlabel('weight', fontsize=18)
plt.ylabel('cost', fontsize=18)
plt.show()    
```

그래프를 통해 기울기(가중치)가 1일 때(순간 기울기가 0인 지점) cost값이 가장 작음을 알 수 있다

### 경사 하강법(gradient descent)
- 그래프에서 오차를 비교하여 가장 작은 방향으로 이동시키는 방법
- 기울기(가중치)를 변화시켜 cost값이 가장 작을 때의 기울기(가중치)를 찾는 작업
- 학습률(learning rate): 기울기를 얼마나 변화시킬지 정하는 값, 적절한 값을 찾아야 한다

#### 경사 하강법 계산

```python
# 순간 기울기 계산
def cal_gradient(x, y, w):
    hx = w * x
    g = 2 * np.mean((hx - y) * x)
    
    return g
```

```python
# 경사 하강법
def gradient_descent(x, y, w):
    # 200번 반복
    for i in range(200):
        # cost값
        c = cost(x, y, w)
        # 순간 기울기
        g = cal_gradient(x, y, w)
        # 기울기(가중치) 업데이트 
        # 0.1: learning rate(학습률)
        w -= 0.1 * g
    # 최종 기울기 출력
    print('w: ', w)
```

```python
# 기울기 초기값 10
gradient_descent(x_data, y_data, 10)
```

경사 하강법을 통해 구한 기울기가 1로 위에서 구한 값과 동일함을 확인 할 수 있다.

### scipy 라이브러리 이용

```python
import scipy.stats as st
```

```python
result = st.linregress(x_data, y_data)
result
```

```python
print('선형 회기 기울기: ', result.slope)
print('절편: ', result.intercept)
print('상관계수: ', result.rvalue)
print('p-value: ', result.pvalue)
print('표준편차: ', result.stderr)
```
