### 정규분포

**정규분포(normal distribution)** 혹은 가우스 정규분포(Gaussian normal distribution)라는 분포는 자연 현상에서 나타나는 숫자를 확률 모형으로 모형화할 때 많이 사용한다.

정규분포는 평균 $\mu$와 분산 $\sigma^2$이라는 두 모수만으로 정의되며 확률밀도함수(pdf: probability density function)는 다음과 같은 수식으로 표현된다.

$$
\begin{align}
\mathcal{N}(x; \mu, \sigma^2) = \dfrac{1}{\sqrt{2\pi\sigma^2}} \exp \left(-\dfrac{(x-\mu)^2}{2\sigma^2}\right)
\tag{1}
\end{align}
$$

분산의 역수를 정밀도(precision) $\beta$라고 부르기도 한다.

$$
\begin{align}
\beta = \dfrac{1}{\sigma^2}
\tag{2}
\end{align}
$$

정규분포 중에서도 평균이 0이고 분산이 1인 ($\mu=0$, $\sigma^2=1$) 정규분포를 **표준정규분포(standard normal distribution)**라고 한다.

정규분포의 확률밀도함수는 다음과 같은 성질을 가진다.

* $x=\mu$일 때 확률밀도가 최대가 된다.
* $x=\infty$로 다가가거나 $x=-\infty$로 다가갈수록 확률밀도가 작아진다.

### 가설 검정
- 모집단에서 추출된 표본을 사용하여 통계적 가설의 옳고 그름을 판단하는 것
- 연구자 자신이 주장하고자 하는 가설을 대립가설(H1), 그 반대의 가설을 귀무가설(H0, 영가설)이라고 부른다
- 대립가설은 보통 '효과가 있다, 연관 되어 있다' 등으로 기술이 되며 우리가 관심을 갖는 가설, 반대로 귀무가설(영가설)은 '효과가 없다, 연관 없다' 등으로 기술 되고 일반적으로 부정하고자 하는 사실들을 귀무가설에 기술한다.
- 유의 확률(p-value)은 귀무가설을 가정하였을 때 표본 이상으로 극단적인 결과를 얻을 확률이다.
- 통계값인 p-value가 0.05보다 작게되면, 귀무가설(영가설)을 기각하고 대립가설 채택, 0.05보다 크면 귀무가설 채택

### Z 검정
- 모집단의 통계치와 표본의 통계치를 비교하는 방법
- 귀무 가설하에서 테스트(검정) 통계량의 분포를 정규 분포로 근사할 수있는 통계 검정이다
- 모집단 분산 σ2를 이미 알고있는 분포의 평균을 테스트한다
- 표본 크기가 크거나 모집단 분산이 알려진 경우 대략적인 Z-테스트로 많은 통계 검정을 편리하게 수행할 수 있다. 모집단 분산을 알 수 없고(따라서 표본 자체에서 추정해야 함), 표본 크기가 크지 않은 경우(n <30), T-테스트가 더 적합 할 수 있다.

### T 검정
- 검정통계량이 귀무가설 하에서 t-분포를 따르는 통계적 가설 검정법이다.
- 일반적으로 검정통계량이 정규 분포를 따르며 분포와 관련된 스케일링 변숫값들이 알려진 경우에 사용한다.
- 모집단의 분산(표준편차)가 주어지지 않을 때 표본의 표준편차를 이용한다.

#### 표준정규분포 확률밀도함수 계산

```python
# 표준정규분포: 평균=0, 분산=1
# -5 ~ 5 100개 일정 간격으로 추출
x = np.linspace(-5, 5, 100)
# 확률밀도함수 계산
y = 1/np.sqrt(2 * np.pi) * np.exp(-(x ** 2 / 2))

# 표준정규분포 그리기
plt.plot(x, y)
plt.show()
```

#### 표준정규분포 scipy 라이브러리 사용

```python
# 표준정규분포
rv = stats.norm(0, 1)
x = np.linspace(-5, 5, 100)
# 확률밀도함수
y = rv.pdf(x)

# 표준정규분포 그리기
plt.plot(x, y)
plt.show()
```

### 누적분포함수
- 누적 분포 함수(cumulative distribution function, cdf)는 주어진 확률 변수가 특정 값보다 작거나 같은 확률을 나타내는 함수이다.

#### 누적분포함수 scipy 라이브러리 사용

```python
# 표준정규분포
rv = stats.norm(0, 1)
x = np.linspace(-5, 5, 100)
# 누적분포함수
y = rv.cdf(x)

# 누적분포함수 그리기 
plt.plot(x, y)
plt.show()
```

```python
# 표준정규분포에서 왼쪽부터 x=0까지 면적의 합
# 전체 면적 = 1
# 전체 면적의 절반
rv.cdf(0)
```

```python
# 표준정규분포 그래프에서 왼쪽부터 x=1까지의 면적의 합
rv.cdf(1)
```

### 표준화(Standardization)
- 표준화 이유
    - 표준화를 하면 평균이 0 표준편차가 1이 되기 때문에 서로 다른 통계 데이터들을 비교하기 용이해진다.
- z = 편차 / 표준편차

#### 표준화 계산

```python
# 0 <= x < 30, 6행 5열
data = np.random.randint(30, size=(6, 5))
data
```

```python
# 표준화
# 컬럼별 계산 
data_standard = (data - np.mean(data, axis=0)) / np.std(data, axis=0)
data_standard
```

#### 표준화 scipy 라이브러리 사용 

```python
stats.zscore(data)
```

#### 표준화 sklearn 라이브러리 사용

```python
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
result = st.fit_transform(data)
result
```

### 연습 문제 3)
- 어느 회사에서 종업원들의 근무기간을 조사했는데, 평균이 11년이고 표준편차가 4년인 정규분포를 따른다고 한다.
- 이 회사에서 14년 초과 근무한 종업원의 비율을 구하시오.

방법 1)

```python
# 평균 11, 표준편차 4인 정규분포
rv = stats.norm(11, 4)

# 전체에서 14년 미만 빼기 
1 - rv.cdf(14)
```

방법 2)

- 표준정규분포표를 보고 값을 구한다
- P(x > 14) -> p(z > (14 - 11) / 4) = p(z > 0.75) = 1 - p(z <= 0.75)
- 1 - 0.7734 = 0.2266

### 연습 문제 4)
- 어느 고등학교 3학년 학생들의 수학 성적은 평균이 70점이고 분산이 64점인 정규분포를 따른다고 한다.
- 점수가 80점 이상이고 90점 이하일 확률을 구하시오.

방법 1)

```python
rv = stats.norm(70, 8)
rv.cdf(90) - rv.cdf(80)
```

방법 2)

- 표준정규분포표를 보고 값을 구한다
- P(80 <= x <= 90) -> p((80 - 70) / 8 <= z <= (90 - 70) / 8) = p(1.25 <= z <= 2.5) = p(z <= 2.5) - p(z <= 1.25)

방법 3)

```python
# 표준정규분포
rv = stats.norm(0, 1)
z1 = (90 - 70) / 8
z2 = (80 - 70)  / 8
rv.cdf(z1) - rv.cdf(z2)
```
